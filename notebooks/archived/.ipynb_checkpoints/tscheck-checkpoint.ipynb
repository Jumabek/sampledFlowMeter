```python
filename  = '/data/juma/data/ids18/CSVs/WS/Friday-16-02-2018/Friday-16-02-2018.pcap_Flow.csv'
df = pd.read_csv(filename, usecols=['Flow ID','Src IP','Dst IP','TimestampMCS'])

MCS_IN_SEC = 1000000

g = df.groupby(['Flow ID'])
flowLC = g.agg({'Flow ID':'count','Src IP':np.max,'Dst IP':np.min,'TimestampMCS':[np.min,np.max]})


flowLC['DurationMCS'] = (flowLC[('TimestampMCS','amax')] - flowLC[('TimestampMCS','amin')])
flowLC['DurationMCS'] = flowLC['DurationMCS'].astype(float)
flowLC['DurationSEC'] = flowLC['DurationMCS']/MCS_IN_SEC
flowLC = flowLC.sort_values(by=['DurationMCS'],ascending=False)

flowLC.to_csv(filename.replace('.pcap_Flow.csv','.Flow_Duration_small.csv'))
```
