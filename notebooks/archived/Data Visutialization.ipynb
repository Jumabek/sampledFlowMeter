```python
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import numpy as np
%matplotlib inline
data = pd.read_csv('/home/juma/data/net_intrusion/CIC-IDS-2018/CSVs/whole_l/Friday-16-02-2018_TrafficForML_CICFlowMeter.csv',encoding='utf-8')

```

```python
def plot_2d_space(X, y, custom=True, classes=['Benign'], label='Classes'):   
    colors = ['b', 'g','r','c','m','y','b','w','b', 'g','r','c','m','y','b','w']
    markers = ['o','^','o','o','o','o','o', 's','s','s','s','s','s']
    for i,v in enumerate(sorted(np.unique(y))):
        if custom:
            if v not in classes:
                continue
        plt.scatter(
            X[y==v, 0],
            X[y==v, 1],
            c=colors[i], label=v, marker=markers[i]
        )
    plt.title(label)
    plt.legend(loc='upper right')
    plt.show()
```
