```python
import pandas as pd
from os.path import join
from collections import defaultdict
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
markers = ['>','*','<','o','^','X','D','8','h','d','>','*','<','o']
def matplotlib_setup():
    figure(figsize=(25,12))
    font = {'size'   : 42}
    plt.rc('font', **font)
    plt.rcParams["axes.linewidth"]  = 2.5    
    plt.grid(linewidth=3)
```

```python
roots = ['/data/juma/data/ids18/CSVs_r_1.0/SR_10/RPS_SI_10_l',
         '/data/juma/data/ids18/CSVs_r_1.0/SR_10/RPS_SI_10_l',
         '/data/juma/data/ids18/CSVs_r_1.0/SR_10/RPS_SI_10_l',
         '/data/juma/data/ids18/CSVs_r_1.0/SR_10/RPS_SI_10_l',
        ]
rps_avg = []
rps_avg_benign = []
for root in roots:
    df = pd.read_csv(join(root,'observation_rate.csv'))
    rps_avg.append(df[df['Label']=='Macro average Observation Rate']['Observation rate'].values[0])
    rps_avg_benign.append(df[df['Label']=='Benign']['Observation rate'].values[0])
print(rps_avg)
print(rps_avg_benign)
```

```python
roots = ['/data/juma/data/ids18/sampler_memory_archive/CSVs_r_1.0_m_0.05/SR_10/SFS_SI_9.77_l',
        '/data/juma/data/ids18/sampler_memory_archive/CSVs_r_1.0_m_0.2/SR_10/SFS_SI_9.77_l',
         '/data/juma/data/ids18/sampler_memory_archive/CSVs_r_1.0_m_0.6/SR_10/SFS_SI_9.77_l',
         '/data/juma/data/ids18/CSVs_r_1.0_m_1.0/SR_10/SFS_SI_9.77_l'
        ]

sfs_avg = []
sfs_avg_benign = []
for root in roots:
    df = pd.read_csv(join(root,'observation_rate.csv'))
    #print(df.head())
    sfs_avg.append(df[df['Label']=='Macro average Observation Rate']['Observation rate'].values[0])
    sfs_avg_benign.append(df[df['Label']=='Benign']['Observation rate'].values[0])
print(sfs_avg)
print(sfs_avg_benign)
```

```python
roots = [    
    '/data/juma/data/ids18/sampler_memory_archive/CSVs_r_1.0_m_0.05/SR_10/FFS_(8,16,4)_l',
    '/data/juma/data/ids18/sampler_memory_archive/CSVs_r_1.0_m_0.2/SR_10/FFS_(8,16,4)_l',
    '/data/juma/data/ids18/sampler_memory_archive/CSVs_r_1.0_m_0.6/SR_10/FFS_(8,16,4)_l',
    '/data/juma/data/ids18/CSVs_r_1.0_m_1.0/SR_10/FFS_(8,16,4)_l'
        ]

ffs_avg = []
ffs_avg_benign = []
for root in roots:
    df = pd.read_csv(join(root,'observation_rate.csv'))
    #print(df.head())
    ffs_avg.append(df[df['Label']=='Macro average Observation Rate']['Observation rate'].values[0])
    ffs_avg_benign.append(df[df['Label']=='Benign']['Observation rate'].values[0])
print(ffs_avg)
print(ffs_avg_benign)
```

```python
roots = [
        '/data/juma/data/ids18/sampler_memory_archive/CSVs_r_1.0_m_0.05/SR_10/SGS_e_0.00285_l',
         '/data/juma/data/ids18/sampler_memory_archive/CSVs_r_1.0_m_0.2/SR_10/SGS_e_0.0108_l',
        #'/data/juma/data/ids18/CSVs_r_1.0_m_0.2/SR_10/SGS_e_0.0107',
        '/data/juma/data/ids18/sampler_memory_archive/CSVs_r_1.0_m_0.6/SR_10/SGS_e_0.0305_l',
        '/data/juma/data/ids18/CSVs_r_1.0_m_1.0/SR_10/SGS_e_0.05_l',
]

sgs_avg = []
sgs_avg_benign = []
for root in roots:
    df = pd.read_csv(join(root,'observation_rate.csv'))
    #print(df.head())
    sgs_avg.append(df[df['Label']=='Macro average Observation Rate']['Observation rate'].values[0])
    sgs_avg_benign.append(df[df['Label']=='Benign']['Observation rate'].values[0])
print(sgs_avg)
print(sgs_avg_benign)
```

```python
matplotlib_setup()
plt.plot(rps_avg_benign,label = 'SRS - Simple Random S.',linewidth=7,marker=markers[0],markersize=30,markeredgewidth=8,markerfacecolor='none')
plt.plot(sfs_avg_benign,label = 'SFS - SketchFlow S.',linewidth=7,marker=markers[1],markersize=30,markeredgewidth=8,markerfacecolor='none')
plt.plot(sgs_avg_benign,label = 'SGS - Sketch Guided S.',linewidth=7,marker=markers[2],markersize=30,markeredgewidth=8,markerfacecolor='none')
plt.plot(ffs_avg_benign,label = 'FFS - Fast Filtered S.',linewidth=7,marker=markers[3],markersize=30,markeredgewidth=8,markerfacecolor='none')
plt.legend(ncol=2)
plt.ylabel('Benign Flow Obsr. Rate')
plt.xlabel("Sampler Memory (Mb)")
plt.yticks(np.arange(40,110,10),[str(i)+'%' for i in np.arange(40,110,10)])
plt.ylim([40,100])
plt.xticks(np.arange(4),['0.05 Mb','0.2 Mb','0.6 Mb','1.0 Mb'])
plt.savefig('/data/juma/data/ids18/ben_observation_rate_across_sampler_memory.png')
plt.show()
```
