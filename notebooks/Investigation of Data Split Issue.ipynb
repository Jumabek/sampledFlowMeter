```python
import pandas as pd
from glob import glob
from os.path import join
import ntpath
```

```python
findex=0
csv_regex1 = 'CSVs_r_1.0_m_1.0'
datadir = '/data/juma/data/ids18/{}/SR_1.0/SRS_SI_100_l/'
regex = '*Meter.csv'
for i, fn1 in enumerate(glob(join(datadir.format(csv_regex1),regex))): 
    print('-----------------------')
    print(ntpath.basename(fn1))
    print('#records, #flows')
    for csv_regex in ['CSVs_r_1.0_m_1.0','CSVs_r_0.1_m_1.0','CSVs_r_0.01_m_1.0','CSVs_r_0.001_m_1.0']:
        fn = fn1.replace(csv_regex1,csv_regex)
        df = pd.read_csv(fn,usecols=['Flow ID'])
        print(df.shape[0],df['Flow ID'].nunique())
```

```python
findex=0
csv_regex1 = 'CSVs_r_1.0_m_1.0'
datadir = '/data/juma/data/ids18/{}/SR_1.0/SRS_SI_100_l/'
regex = '*Meter.csv'
for i, fn1 in enumerate(glob(join(datadir.format(csv_regex1),regex))): 
    print('-----------------------')
    print(ntpath.basename(fn1))
    print('#records, #flows')
    for csv_regex in ['CSVs_r_1.0_m_1.0','CSVs_r_0.1_m_1.0','CSVs_r_0.01_m_1.0','CSVs_r_0.001_m_1.0']:
        fn = fn1.replace(csv_regex1,csv_regex)
        df = pd.read_csv(fn,usecols=['Flow ID','Label'])
        df = df[df['Label']!='Benign']
        print(df.shape[0],df['Flow ID'].nunique())
```
