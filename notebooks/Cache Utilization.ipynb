```python
import pandas as pd
from glob import glob
import numpy as np
import ntpath

from datetime import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from os.path import join
%matplotlib inline
```

```python
def matplotlib_setup():
    figure(figsize=(18,6))
    font = {'size'   : 30}
    plt.rc('font', **font)
    plt.rcParams["axes.linewidth"]  = 2.5
    plt.grid(linewidth=3,axis='y', color='grey')
    
    CB91_Blue = '#2CBDFE'
    CB91_Green = '#47DBCD'
    CB91_Pink = '#F3A0F2'
    CB91_Purple = '#9D2EC5'
    CB91_Violet = '#661D98'
    CB91_Amber = '#F5B14C'
    color_list = [CB91_Blue, CB91_Pink, CB91_Green, CB91_Amber,
              CB91_Purple, CB91_Violet]
    plt.rcParams['axes.prop_cycle'] = plt.cycler(color=color_list)
    #plt.yticks(np.arange(0,121,20),np.arange(0,121,20) )
    plt.ylabel('#WSAF in Flow Cache')
    
patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*" ]
```

```python
def convert_to_datetime(arr):
    fmt = '%H:%M'
    offset = timedelta(hours=11,minutes=36)
    arr_str = []
    for t in arr:
        dt = datetime.fromtimestamp(t/1000000.)
        dt = dt-offset
        arr_str.append(dt.strftime(fmt))
    return arr_str
```

```python
def put_x_ticks(df,plt):
    #put time labels on x-axis
    begin_time = df['Timestamp'].min()
    end_time = df['Timestamp'].max()  
    time_labels = np.linspace(begin_time,end_time,6)
    lindices = convert_to_datetime(time_labels)
    plt.xticks(time_labels,lindices,fontsize=35)
```

```python
SR = 'SR_10.0'
sdirs = [
    '/data/juma/data/ids18/CSVs_r_1.0_m_1.0/{}/SRS_SI_10/'.format(SR),
    '/data/juma/data/ids18/CSVs_r_1.0_m_1.0/{}/SGS_e_0.05/'.format(SR),
    '/data/juma/data/ids18/CSVs_r_1.0_m_1.0/{}/SFS_SI_9.77/'.format(SR),
    '/data/juma/data/ids18/CSVs_r_1.0_m_1.0/{}/FFS_(8,16,4)/'.format(SR),
]
print(join(sdirs[0],'*','MemoryUsage_FI_1_Flow.csv'))
for fn0 in glob(join(sdirs[0],'*','MemoryUsage_FI_1_Flow.csv')):
    print(fn0)
    matplotlib_setup()
    for dtr in sdirs:
        fn = fn0.replace(sdirs[0],dtr)
        #print(fn)
        df = pd.read_csv(fn)
        sdir = ntpath.basename(ntpath.split(ntpath.split(fn)[0])[0])
        plt.plot(df['Timestamp'].values, df['#WSAF'].values,linewidth=3, label=sdir)
        name_of_the_day = ntpath.basename(ntpath.split(fn)[0])


    put_x_ticks(df,plt)    
    plt.xlabel(name_of_the_day)
    plt.legend(ncol=2)
    plt.tight_layout()
    plt.savefig(join(dtr,'cache_util_'+name_of_the_day+'.png'))    
    plt.show()
    print(fn)
```
